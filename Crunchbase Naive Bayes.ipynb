```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16,9)
plt.style.use('ggplot')
```

# Upload and Preprocess Data

```python
df1 = pd.read_csv('crunchbase_aerospace.csv')
df2 = pd.read_csv('crunchbase_ai.csv')
df3 = pd.read_csv('crunchbase_biotech.csv')
df4 = pd.read_csv('crunchbase_energy.csv')
df5 = pd.read_csv('crunchbase_neurotech.csv')

df1
```

```python
df1['Industries'] = 'Aerospace'
df2['Industries'] = 'Artificial Intelligence'
df3['Industries'] = 'Biotechnology'
df4['Industries'] = 'Energy'
df5['Industries'] = 'Neurotech'
```

```python
# stick together dataframes
data = pd.concat([df1,df2,df3,df4,df5])
data = data.reset_index()
data
```

```python
data = data.drop(columns=['index','Organization Name URL', 'Operating Status', 'Company Type', 'Headquarters Location', 'Description', 'IPO Status', 'Estimated Revenue Range', 'Last Funding Type', 'Total Funding Amount Currency', 'Total Funding Amount', 'Full Description', 'Founded Date Precision', 'Website'])
```

```python
data = data.dropna(subset=['Number of Employees', 'Total Funding Amount Currency (in USD)', 'Founded Date'])
```

```python
data = data.drop_duplicates(subset=['Organization Name'])
```

```python
def to_year(date):
    date = date[:4]
    age = 2022-int(date)
    return age

for i in data.index:
    if type(data['Founded Date'][i]) == str:
        data['Founded Date'][i] = to_year(data['Founded Date'][i])
    elif type(data['Founded Date'][i]) == float:
        print(data['Founded Date'][i])
```

```python
data = data.rename({'Founded Date': 'Age'}, axis=1)
data
```

```python
list(data['Number of Employees'].unique())
```

```python
employee_counts = list(data['Number of Employees'].unique())
rounded_counts = [10001,101,1001,251,11,51,1,501,5001]
print(len(employee_counts))
print(len(rounded_counts))
print(employee_counts)
```

```python
employee_counts = ['11-50', '101-250', '51-100', '251-500', '1001-5000', '1-10', '10001+', '501-1000', '5001-10000']
rounded_counts = [10001,101,1001,251,11,51,1,501,5001]

def estimate_employees(employees):
    for i in range(len(employee_counts)):
        if employees == employee_counts[i]:
            employees = rounded_counts[i]
    return employees

for i in data.index:
    data['Number of Employees'][i] = estimate_employees(data['Number of Employees'][i])
```

```python
data = data.drop(['Organization Name'], axis=1)
data
```

```python
def get_unique_values_with_dictionary(values):
    new_values = {}

    for value in values:
        if value in new_values.keys():
            new_values[value] += 1
        else:
            new_values[value] = 1
    
    print(new_values)
    return new_values.keys()

get_unique_values_with_dictionary(list(data['Industries']))
```

# Build Model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
X = data[["Number of Employees","Total Funding Amount Currency (in USD)","Age"]]
Y = data[["Industries"]]
Xtrain, Xtest, ytrain, ytest = train_test_split(X, Y, random_state=1000)
```

```python
model = GaussianNB()
model.fit(Xtrain, ytrain)
```

```python
# classes in sample
model.classes_
```

```python
# count of each class in sample
model.class_count_
```

```python
# probability of each class in sample
model.class_prior_
```

# Predicting Test Data

```python
ypred = model.predict(Xtest)
accuracy = accuracy_score(ytest,ypred)
report = classification_report(ypred, ytest)
cm = confusion_matrix(ytest, ypred)
```

```python
print("Classification report:")
print("Accuracy: ",accuracy)
print(report)
print("Confusion matrix:")
print(cm)
```
